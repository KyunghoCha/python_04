```python
print('2025-11-04, 21114073, 차경호\n')

def nSum(n: int = 0) -> int:
    result = 0
    for i in range(1, n + 1):
        result += i

    return result

number = int(input("양의 정수를 입력하세요>> "))
result = nSum(number)
print(f'1부터 {number}까지의 합 = {result}')
```

```python
print('2025-11-04, 21114073, 차경호\n')

def u_nSum(start: int = 1, end: int = 100) -> int:
    result = 0
    for i in range(start, end + 1):
        result += i

    return result


user_start = int(input('시작 정수를 입력하세요>>'))
user_end = int(input('마지막 정수를 입력하세요>>'))
result = u_nSum(user_start, user_end)
print(f'{user_start}부터 {user_end}까지의 합 = {result}')
```

```python
print('2025-11-04, 21114073, 차경호\n')

def area(w: int, h: int) -> int:
    return w * h

def input_pos(msg: str) -> int:
    while True:
        value = int(input(msg))
        if value > 0:
            return value
        else:
            print('양수만 입력하시오.')

width = input_pos('사각형의 가로: ')
height = input_pos('사각형의 세로: ')

result = area(width, height)
print('면적=', result)
```

```python
print('2025-11-04, 21114073, 차경호\n')

def fact(n: int) -> int:
    if n == 0:
        return 1
    return n * fact(n - 1)

def input_pos(msg: str) -> int:
    while True:
        value = int(input(msg))
        if value > 0:
            return value
        else:
            print('양수만 입력하시오.')

n = input_pos('양수를 입력하시오')
print(fact(n))
```

```python
print('2025-11-04, 21114073, 차경호\n')

def analyse_string(text: str) -> tuple:
    length = len(text)
    words = len(text.split())
    return length, words

input_text = input('문자열을 입력하시오.')

result_length, result_words = analyse_string(input_text)

print(f'문자열 길이: {result_length}')
print(f'단거 개수" {result_words}')
```
