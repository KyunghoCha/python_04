```python
print('2025-11-11, 21114073, 차경호\n')

def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n - 1)

n = int(input('정수를 입력하세요>> '))
print(f'{n}! = {fact(n)}')
```

```python
import os

print('2025-11-11, 21114073, 차경호\n')

def calcDirSize(name: str) -> int:
    totalSize = 0

    if os.path.isfile(name):
        totalSize += os.path.getsize(name)
    else:
        dirList = os.listdir(name)
        for subDir in dirList:
            totalSize += calcDirSize(os.path.join(name, subDir))

    return totalSize

name = input('디렉토리 이름을 입력하시오: ')
print(calcDirSize(name))
```

```python
print('2025-11-11, 21114073, 차경호\n')

def calculate_sum(integer_list: list) -> int:
    result_sum = 0

    for number in integer_list:
        result_sum += number
    return result_sum

example_list = [1, 2, 3, 4, 5]

result_sum = calculate_sum(example_list)

print(f'리스트 {example_list}의 합: {result_sum}')
```

```python
print('2025-11-11, 21114073, 차경호\n')

def readList() -> list:
    nlist = []
    flag = True

    while flag:
        number = int(input('숫자를 입력하시오: '))
        if number < 0:
            flag = False
        else:
            nlist.append(number)
    return nlist

def processList(nlist: list) -> list:
    nlist.sort()
    return nlist

def printList(nlist: list) -> None:
    for i in nlist:
        print(f'점수= {i}')

def main():
    nlist = readList()
    processList(nlist)
    printList(nlist)

if __name__ == '__main__':
    main()
```

```python
print('2025-11-11, 21114073, 차경호\n')

def find_max(integer_list: list) -> int:
    max = 0

    for number in integer_list:
        if number > max:
            max = number
    return max

example_list = [1, 2, 3, 4, 5]

result_sum = find_max(example_list)

print(f'리스트 {example_list}의 최대 값 : {result_sum}')
```

```python
print('2025-11-11, 21114073, 차경호\n')

STUDENTS = 5

scores = []
scoreSum = 0
highScoreStudents = 0

for i in range(STUDENTS):
    value = int(input('성적을 입력하세요>> '))
    scores.append(value)
    scoreSum += value

    if scores[i] > 80:
        highScoreStudents += 1

scoreAvg = scoreSum / len(scores)
print(f'\n점수의 평균은 {scoreAvg} 입니다.')
print(f'평균이 80점 이상인 학생은 {highScoreStudents} 명 입니다.')
```

```python
print('2025-11-11, 21114073, 차경호\n')

friendNames = []

while True:
    name = input('찬구의 이름을 입력하시오(종료시에는 엔터키): ')

    if name == '':
        break

    friendNames.append(name)

print('친구들의 이름:')

for name in friendNames:
    print(name, end=', ')
```

```python
print('2025-11-11, 21114073, 차경호\n')

keyword_list = []
user_input = ''

while user_input != '끝':
    user_input = input('친구 이름 >>')
    if user_input != '끝':
        if user_input not in keyword_list:
            keyword_list.append(user_input)
        else:
            print(f'{user_input}은 이미 목록에 포함되어 있습니다.')

result = sorted(keyword_list)

keyword_count = len(keyword_list)
print('입력된 친구 수 :', keyword_count)
print('친구 목록 :', keyword_list)
print('이름 순 정렬', result)
```
