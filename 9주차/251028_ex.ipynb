```python
start = int(input('시작 값을 입력하세요: '))
end = int(input('마지막 값을 입력하세요: '))

def get_sum(start: int = 0, end: int = 10) -> int:
    sum = 0
    for i in range(start, end + 1):
        sum += i
    return sum

print(f'{start}부터 {end}까지의 합계 = {get_sum(start, end)}')
```

```python
print('2025-10-28, 21114073, 차경호\n')

def is_prime(n: int) -> bool:
    if n < 2: return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

n = int(input('정수를 입력하세요: '))

print(f'정수 {n}은 소수입니까? {is_prime(n)}')
```

```python
print('2025-10-28, 21114073, 차경호\n')

def is_prime(n: int = 0) -> bool:
    if n < 2: return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print('1부터 100까지의 소수')
for number in range(1, 101):
    if is_prime(number):
        print(number, end=' ')
```
