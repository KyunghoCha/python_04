```python
print('2025-10-14, 21114073, 차경호\n')

score = int(input('점수를 입력하세요.'))

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f'당신의 성적 {score}은 ', grade, '입니다.')
```

```python
print('2025-10-14, 21114073, 차경호\n')

k_number = ' 일이삼사오육칠팔구'
k_dec = ' 십백천만'
# k_dec_ex = ' 만억조경해'

number = int(input('1부터 999까지의 정수를 입력해 주세요.'))
num_digits = len(str(number)) - 1

print('영' if number == 0 else ''.join(
    [k_number[(number // (10 ** i)) % 10] + k_dec[i] for i in range(num_digits, -1, -1)]))
```

```python
from random import randint

answer = randint(1, 100)
print('숫자 게임에 오신 것을 환영합니다.')
g = input('숫자를 맞춰 보세요.')
guess = int(g)

if guess == answer:
    print('사용자가 이겼습니다.')
elif guess > answer:
    print('너무 큼!')
elif guess < answer:
    print('너무 작음!')
print('게임 종료')
```

```python
print('2025-10-14, 21114073, 차경호\n')

tries = 0
number = random.randint(0, 100)
while tries < 10:
    guess = int(input('숫자를 맞춰보세요.'))
    tries += 1
    if guess < number:
        print('낮음')
    elif guess > number:
        print('높음')
    else:
        break

if guess == number:
    print('축하합니다. 시도횟수=', tries)
else:
    print('정답은', number)
```

```python
import turtle

t = turtle.Turtle()  #터틀 객체 생성
t.width(3)
#신호 입력 받기
signal = input("(red:r, yellow.y, green:g): ")
#신호에 따라 원 그리기
if signal == "r":
    t.penup()
    t.goto(-200, -100)
    t.pendown()
    t.color("red")
    t.begin_fill()
    t.circle(100)
    t.end_fill()
    t.color("green")
    t.penup()
    t.goto(0, -100)
    t.pendown()
    t.circle(100)
    t.color("yellow")
    t.penup()
    t.goto(200, -100)
    t.pendown()
    t.circle(100)
elif signal == "y":
    t.penup()
    t.goto(-200, -100)
    t.pendown()
    t.color("red")
    t.circle(100)
    t.color("green")
    t.penup()
    t.goto(0, -100)
    t.pendown()
    t.circle(100)
    t.color("yellow")
    t.penup()
    t.goto(200, -100)
    t.pendown()
    t.begin_fill()
    t.circle(100)
    t.end_fill()
elif signal == "g":
    t.penup()
    t.goto(-200, -100)
    t.pendown()
    t.color("red")
    t.circle(100)
    t.color("green")
    t.penup()
    t.goto(0, -100)
    t.pendown()
    t.begin_fill()
    t.circle(100)
    t.end_fill()
    t.color("yellow")
    t.penup()
    t.goto(200, -100)
    t.pendown()
    t.circle(100)
else:
    print("지원되지 않는 신호입니다.")
#터틀 그래픽 종료
turtle.done()
```
